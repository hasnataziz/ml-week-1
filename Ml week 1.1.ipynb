```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
Iris = pd.read_csv("Iris.csv")
Iris = Iris.drop(columns=["Id"])
#to show the relationship between the features
#for individual
#sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
#sns.scatterplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
#sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', hue='Species')
#sns.scatterplot(data=df, x='SepalWidthCm', y='PetalWidthCm', hue='Species')


# Load the dataset





# If you want to create individual scatter plots, you can do it like this:
# Scatter plot between SepalLengthCm and SepalWidthCm
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Scatter plot between Sepal Length and Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()

# Scatter plot between PetalLengthCm and PetalWidthCm
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Iris, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
plt.title('Scatter plot between Petal Length and Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()

# Scatter plot between SepalLengthCm and PetalLengthCm
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Iris, x='SepalLengthCm', y='PetalLengthCm', hue='Species')
plt.title('Scatter plot between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()

# Scatter plot between SepalWidthCm and PetalWidthCm
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Iris, x='SepalWidthCm', y='PetalWidthCm', hue='Species')
plt.title('Scatter plot between Sepal Width and Petal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```
